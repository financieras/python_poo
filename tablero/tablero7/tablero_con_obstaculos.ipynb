# Tablero con obstáculos
Implementar en Python con POO un algoritmo de Backtracking para resolver un laberinto. Se debe implementar en una matriz de 8x8 y de forma aleatoria se deben agregar obstáculos en el tablero. El recorrido siempre inicia en la esquina superior izquierda y debe terminar en la esquina inferior derecha. La primera iteración inicia con 2 obstáculos y cada vez que se llega al objetivo inicia de nuevo agregando 2 obstáculos más. Se debe mostrar paso a paso las decisiones que va tomando el algoritmo de Backtracking para intentar llegar al objetivo. Se debe poder visualizar la ruta completa que lleva en cada paso. Los obstáculos se van incrementando hasta el momento que no se pueda poner más obstáculos para permitir que haya al menos una ruta para llegar al objetivo.

Para implementar el algoritmo de Backtracking en Python con POO para resolver un laberinto de 8x8 con obstáculos aleatorios, puedes seguir los siguientes pasos:

1. Crear una clase Laberinto que contenga la matriz de 8x8 y los obstáculos aleatorios.
2. Crear una clase Nodo que represente cada celda del laberinto y que contenga información sobre su posición, si es un obstáculo o no y sus vecinos.
3. Crear una clase Backtracking que contenga el algoritmo de Backtracking y que reciba como parámetros el nodo actual y el nodo objetivo.
4. Implementar el algoritmo de Backtracking en la clase Backtracking para encontrar la solución óptima.
5. Mostrar paso a paso las decisiones que va tomando el algoritmo de Backtracking para intentar llegar al objetivo.
6. Visualizar la ruta completa que lleva en cada paso.

```python
import random

class Nodo:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.obstaculo = False
        self.vecinos = []

    def agregar_vecino(self, vecino):
        if vecino not in self.vecinos:
            self.vecinos.append(vecino)
            vecino.agregar_vecino(self)

class Laberinto:
    def __init__(self):
        self.matriz = [[Nodo(x, y) for x in range(8)] for y in range(8)]
        self.agregar_obstaculos()

    def agregar_obstaculos(self):
        for i in range(2):
            x = random.randint(0, 7)
            y = random.randint(0, 7)
            self.matriz[x][y].obstaculo = True

        for i in range(8):
            for j in range(8):
                nodo_actual = self.matriz[i][j]

                if i > 0:
                    nodo_arriba = self.matriz[i - 1][j]
                    nodo_actual.agregar_vecino(nodo_arriba)

                if i < 7:
                    nodo_abajo = self.matriz[i + 1][j]
                    nodo_actual.agregar_vecino(nodo_abajo)

                if j > 0:
                    nodo_izquierda = self.matriz[i][j - 1]
                    nodo_actual.agregar_vecino(nodo_izquierda)

                if j < 7:
                    nodo_derecha = self.matriz[i][j + 1]
                    nodo_actual.agregar_vecino(nodo_derecha)

class Backtracking:
    def __init__(self, laberinto):
        self.laberinto = laberinto
        self.camino = []

    def encontrar_camino(self, nodo_actual, nodo_objetivo):
        if nodo_actual == nodo_objetivo:
            return True

        if nodo_actual.obstaculo:
            return False

        self.camino.append(nodo_actual)

        for vecino in nodo_actual.vecinos:
            if vecino not in self.camino:
                if self.encontrar_camino(vecino, nodo_objetivo):
                    return True

        self.camino.pop()

        return False

laberinto = Laberinto()
backtracking = Backtracking(laberinto)

nodo_inicial = laberinto.matriz[0][0]
nodo_final = laberinto.matriz[7][7]

if backtracking.encontrar_camino(nodo_inicial, nodo_final):
    print("Camino encontrado:")
    for nodo in backtracking.camino:
        print(f"({nodo.x}, {nodo.y})")
else:
    print("No se encontró camino.")
```
